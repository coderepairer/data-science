## Concatenation of Dataframes using concat function

```python
import pandas as pd
df1 = pd.DataFrame(
      {'A':['A0','A1','A2','A3'],
       'B':['B0','B1','B2','B3'],
       'C':['C0','C1','C2','C3'],
       'D':['D0','D1','D2','D3']      
      }, index = {0,1,2,3})

df2 = pd.DataFrame(
      {'A':['A4','A5','A6','A7'],
       'B':['B4','B5','B6','B7'],
       'C':['C4','C5','C6','C7'],
       'D':['D4','D5','D6','D7']      
      }, index = {4,5,6,7})

df3 = pd.DataFrame(
      {'A':['A8','A9','A10','A11'],
       'B':['B8','B9','B10','B11'],
       'C':['C8','C9','C10','C11'],
       'D':['D8','D9','D10','D11']      
      }, index = {8,9,10,11})
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
concat_df = pd.concat([df1,df2,df3],keys=['df1','df2','df3'],axis=0)
concat_df
```

```python
#Fetching dataframe by "key" using loc just like we use iloc to fetch dataframe rows & columns
concat_df.loc['df2']
```

```python
concat_df = pd.concat([df1,df2,df3],keys=['df1','df2','df3'],axis=1)
concat_df
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                        'D': ['D2', 'D3', 'D6', 'D7'],
                        'F': ['F2', 'F3', 'F6', 'F7']},
                       index=[2, 3, 6, 7])
df4
```

```python
concat_df = pd.concat([df1, df4], axis=0, sort=False,join="outer")
concat_df
```

```python
concat_df = pd.concat([df1, df4], axis=0, sort=False,join="inner")
concat_df
```

```python
concat_df = pd.concat([df1, df4], axis=1, sort=False,join="outer")
concat_df
```

```python
concat_df = pd.concat([df1, df4], axis=1, sort=False,join_axes=[df1.index],join="outer")
concat_df
```

```python
concat_df = pd.concat([df1, df4], axis=1, sort=False,join="inner")
concat_df
```

## Merging dataframes using merge function

```python
df_a = pd.DataFrame({
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']})

df_b = pd.DataFrame({
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']})

df_c = pd.DataFrame({
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]})
```

```python
df_a
```

```python
df_b
```

```python
df_c
```

```python
merge_df = pd.merge(df_a,df_c)
merge_df
```

```python
merge_df = pd.merge(df_a,df_c,on="subject_id")
merge_df
```

```python
## Full Outer Join
merge_df = pd.merge(df_a,df_c,how="outer")
merge_df
```

```python
## Inner Join
merge_df = pd.merge(df_a,df_b,on="subject_id",how="inner")
merge_df
```

```python
## Left Outer Join
merge_df = pd.merge(df_a,df_c,how="left")
merge_df
```

```python
## Right Outer Join
merge_df = pd.merge(df_a,df_c,how="right")
merge_df
```

## Exploring Big Mart data using Matplotlib

```python
#Plot to show Mean price per Item Type
import pandas as pd

bm_df = pd.read_csv("../data/bigmart_data.csv")
bm_df=bm_df.dropna(how="any")

bm_df.head()
```

```python
#Plot to show Mean price per Item Type
grouped_df = bm_df.groupby("Item_Type")

grouped_df.first()
```

```python
#Plot to show Mean price per Item Type
price_per_item = grouped_df.Item_MRP.mean()[:10]
price_per_item
```

```python
#Plot to show Mean price per Item Type
import numpy as np
x=price_per_item.index.tolist()
y= price_per_item.values.tolist()

plt.figure(figsize=(18,8))


plt.title("Mean Price per Item Type")
plt.xlabel("Item Type")
plt.ylabel("Mean Price")
plt.xticks(labels=x,ticks=np.arange(len(x)))
plt.plot(x,y)
```

```python
#Bar Plot to show Mean sales by Outlet size
sales_by_outlet_size = bm_df.groupby("Outlet_Size").Item_Outlet_Sales.mean()
sales_by_outlet_size.sort_values(inplace=True)
sales_by_outlet_size
```

```python
#Bar Plot to show Mean sales by Outlet size
x=sales_by_outlet_size.index.tolist()
y=sales_by_outlet_size.values.tolist()

#plt.figure(figsize=(10,8))

plt.title("Mean Sales by Outlet Size")
plt.xlabel("Outlet Size")
plt.ylabel("Mean Sales")
plt.xticks(labels=x,ticks=np.arange(len(x)))
plt.bar(x,y,color=['orange','green','pink'],)
```

```python
#Histogram to show Distribution of Item MRP (Price)

plt.title("Item MRP (Price) Distribution")
plt.xlabel("Item MRP")
plt.ylabel("Frequency")
plt.hist(bm_df['Item_MRP'],bins=10,color=['lightblue']);
```

```python
#BoxPlot to show Distribution of Sales (Outliers)
x=bm_df['Item_Outlet_Sales']
red_circled_outliers=dict(markerfacecolor='r',marker='o')
plt.title("Item Sales Distribution")
plt.boxplot(x.values,labels=['Item Outlet Sales'],flierprops=red_circled_outliers);
```

```python
#Multiple BoxPlot to show Distribution of Item_MRP & Item_Weight (Outliers)
fig,ax=plt.subplots()
x=bm_df[['Item_MRP','Item_Weight']]
red_circled_outliers=dict(markerfacecolor='r',marker='o')
plt.title("Distribution of Item Weight & Item MRP(Price)")
plt.boxplot(x.values,labels=['Item_MRP','Item_Weight'],flierprops=red_circled_outliers);
```

```python
#ViolinPlot to show density of Distribution of Item_MRP & Item_Weight
fig,ax=plt.subplots()
x=bm_df[['Item_MRP','Item_Weight']]
plt.title("Density Distribution of Item Weight & Item MRP(Price)")
plt.xticks(ticks=[1,2],labels=['Item_MRP','Item_Weight'])
plt.violinplot(x.values);
```

```python
#ScatterPlot to show distribution of Item Weight & Item Visibility
plt.figure(figsize=(15,8))
plt.title("Distribution of Item Weight & Item Visibility")
colors=[]
for i in range(100): 
    colors.append(np.random.random_sample())
plt.xlabel('Item_Weight')
plt.ylabel('Item_Visibility')
plt.scatter(bm_df['Item_Weight'][:100],bm_df['Item_Visibility'][:100],s=50,label=['Item_Weight','Item_Visibility'],c=colors);
```

```python
#BubblePlot(like multiple scatterplot) to show distribution of Item MRP, Item Outlet Sales, Item Visibility
plt.figure(figsize=(15,12))
plt.title("Distribution of Item MRP, Item Outlet Sales & Item Visibility")
plt.xlabel('Item_MRP')
plt.ylabel('Item_Outlet_Sales')

colors=[]
for i in range(100): 
    colors.append(np.random.random_sample())
    
    
plt.scatter(bm_df['Item_MRP'][:100],bm_df['Item_Outlet_Sales'][:100],s=bm_df['Item_Visibility'][:100]*1000,c=colors);
```

## Regular Expressions

```python
import re

#Matching pattern at the beginning of the line and printing result group
result = re.match('Analytics',r'Analytics Vidhya is the largest data science community of India')
print(result)

result2 = re.match('largest',r'Analytics Vidhya is the largest data science community of India')
print(result2)

print(result.group())
```

```python
#Searching the occurence of the pattern line and printing result group
result = re.search('founded',r'Andrew NG founded Coursera. He also founded deeplearning.ai')
print(result)
print(result.group())
```

```python
#Finding all the occurences of the pattern line and printing result group
result = re.findall('founded',r'Andrew NG founded Coursera. He also founded deeplearning.ai')
print(result)
```
